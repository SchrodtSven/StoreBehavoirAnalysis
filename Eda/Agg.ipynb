```python
import pandas as pd

month = "2019-Oct"
#month = "2019-Nov"
#month = "2019-Dec"
#month = "2020-Jan"
#month = "2020-Feb"
#month = "2020-Mar"
#month = "2020-Apr"
del df
df = pd.read_csv(f"san_{month}.csv")

fld = 'agg/'

df.head(6)
```

### X per Day Aggregation
- Category
- Brand
- Product ???

```python
subj = "cat_per_day_sales_"
file = f"{fld}{subj}{month}.csv"

cat_per_day_sales = df.groupby(["date", "category_id"])["price"].sum()
cat_per_day_sales.to_csv(file)
cat_per_day_sales.head()
```

```python
subj = "brnd_per_day_sales_"
file = f"{fld}{subj}{month}.csv"

brnd_per_day_sales = df.groupby(["date", "brand"])["price"].sum()
brnd_per_day_sales.to_csv(file)
brnd_per_day_sales.head(2)
```

```python
subj = "prd_per_day_sales_"
file = f"{fld}{subj}{month}.csv"

prd_per_day_sales = df.groupby(["date", "product_id"])["price"].sum()
prd_per_day_sales.to_csv(file)
prd_per_day_sales.head(13)
```
